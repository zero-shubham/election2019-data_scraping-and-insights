```python
import os
import csv
```

```python
wbConstituencies = os.listdir("./constituencyWise2/West_Bengal/")
wbConstituencies
```

```python
path = "./constituencyWise2/West_Bengal/"
starCandDict = dict()
for constituency in wbConstituencies:
    with open(path+constituency, "r") as csvfile:
        reader = csv.reader(csvfile,delimiter=",")
        for row in reader:
            if row[2] == 'All India Trinamool Congress':
                if not starCandDict:
                    starCandDict["name"] = row[1]
                    starCandDict["total_votes"] = int(row[5])
                    starCandDict["constituency"] = constituency
                else:
                    if int(row[5]) > starCandDict["total_votes"]:
                        starCandDict["name"] = row[1]
                        starCandDict["total_votes"] = int(row[5])
                        starCandDict["constituency"] = constituency
            
print(starCandDict)
```

```python
#let's find out star candidate for BJP across India
states = os.listdir("./constituencyWise2")
```

```python
path = "./constituencyWise2/"
starCandDict = dict()
secondStar = None
for state in states:
    constituencies = os.listdir(path+"/"+state)
    for constituency in constituencies:
        with open(path+state+"/"+constituency, "r") as csvfile:
            reader = csv.reader(csvfile,delimiter=",")
            for row in reader:
                if row[2] == 'Bharatiya Janata Party':
                    if not starCandDict:
                        starCandDict["name"] = row[1]
                        starCandDict["total_votes"] = int(row[5])
                        starCandDict["constituency"] = constituency
                    else:
                        if int(row[5]) > starCandDict["total_votes"]:
                            secondStar = starCandDict.copy()
                            starCandDict["name"] = row[1]
                            starCandDict["total_votes"] = int(row[5])
                            starCandDict["constituency"] = constituency
print(starCandDict, "\n",secondStar)
```
